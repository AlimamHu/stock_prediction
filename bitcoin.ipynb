```python
%matplotlib inline
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
bitc=pd.read_csv('BTC-INR.csv')
```

```python
bitc.columns
```

```python
bitc1=bitc.sort_values(by='Date',ascending=False).set_index('Date')
```

```python
bitc1.info()
```

```python
bitc1.describe().T
```

```python
total_missing=bitc1.isna().sum().sum()

total_cells=np.product(bitc1.shape) 


print('missing value in percentage ---> ',total_missing/total_cells*100,'%' ,f'\n total missing values--> {total_missing}\n total cells--> {total_cells}')
```

```python
bitc1.dropna(axis=0,inplace=True)
```
